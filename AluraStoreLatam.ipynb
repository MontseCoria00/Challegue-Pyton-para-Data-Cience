# 4. Productos más y menos vendidos

```python
analisis_productos = {}


for i, tienda_df in enumerate([tienda, tienda2, tienda3, tienda4]):
    nombre_tienda = f"tienda{i+1}"


    contar_productos = tienda_df['Producto'].value_counts()


    analisis_productos[nombre_tienda] = {
        "mas_vendido": contar_productos.index[0],
        "menos_vendido": contar_productos.index[-1]
    }

display(analisis_productos)
```

# Conclusión

Se hizo el análisis de las cuatro tiendas del Señor Juan, con el objetivo de poder ayudarlo a tomar una decisión, tratando de identificar la tienda menos eficiente.

Los ingresos totales de las tiendas, se muestran a continuación :

```python
for i in range(len(Tiendas)):
  print (f'{Tiendas[i]} : {facturacion_Total[i]}')
```

#### Para poder visualizarlo de mejor manera se hizo una gráfica de barras

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.bar(Tiendas, facturacion_Total, color='skyblue')
plt.xlabel('Tiendas')
plt.ylabel('Facturación Total')
plt.title('Facturación Total de Todas las Tiendas')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
print(f'En la grafica obtenida se puede notar que la tienda número uno es la que más factura y la tienda cuatro es la que menos factura, con una diferencia de {facturacion_1 -facturacion_4 }')
```

#### Además es importante profundizar en las ventas de cada tienda, para obtener la tienda de dónde se obtengan más igresos y de manera contraria, es decir:de la que se obtengan menos ventas, para poder observarlo de manera visual se gráfico, como se muestra en seguida

```python
plt.figure(figsize=(10,6))
plt.plot(Tiendas, facturacion_Total, marker='o', linestyle='-', color='blue')
plt.xlabel('Tiendas')
plt.ylabel('Facturación Total')
plt.title('Ventas Totales por Tienda (Gráfico de Líneas)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

De la gráfica obtenida se puede notar que la tienda uno es la que tiene más ventas, con respecto a las tres rerstantes, y la tienda cuatro es la menor en ventas.
Continuando con el  análisis, cabe destacar las valoraciones de las tiendas de parte de los clientes, que nos darán una vision global de popularidad de las mismas.

```python
plt.figure(figsize=(10,6))
plt.scatter(Tiendas, calificación_Total, color='red', s=100) # s is marker size
plt.xlabel('Tiendas')
plt.ylabel('Calificación Promedio')
plt.title('Calificación Promedio de Todas las Tiendas ')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Aquí se tiene un contraste interesante ya que la tiendas más popular es la tres y la menos popular es la uno.
